```python
#1.	Find all of the numbers from 1-1000 that are divisible by 7
l=[i for i in range(1,1000) if i%7==0]
print(l)
```

```python
#2.	Find all of the numbers from 1-1000 that have a 3 in them
l=[i for i in range(1,1000) if '3' in str(i)]
print(l)
```

```python
#3.	Count the number of spaces in a string

st="hello people hii"
l=[s for s in st if s == ' ' ]
print(len(l))
```

```python
#4.	Create a list of all the consonants in the string
#“Yellow Yaks like yelling and yawning and yesturday they yodled while eating yuky yams”
st="Yellow Yaks like yelling and yawning and yesturday they yodled while eating yuky yams"
l=[s for s in st if s  not in "aeiouAEIOU"]
print(l)
```

```python
#5.	Get the index and the value as a tuple for items in the list 
#“hi”, 4, 8.99, ‘apple’, (‘t,b’,’n’). Result would look like (index, value), (index, value)
items = ["hi", 4, 8.99, 'apple', ('t,b','n')]
result = [(index, i) for index,i in enumerate(items)]
print(result)
```

```python
#6.	Find the common numbers in two lists (without using a tuple or set) list_a = 1, 2, 3, 4, list_b = 2, 3, 4, 5
list1=1,2,3,4
list2=2,3,4
l=tuple(s for s in list1 if s in list2)
print(l)
```

```python

#7.	Get only the numbers in a sentence like 
#‘In 1984 there were 13 instances of a protest with over 1000 people attending’
s="In 1984 there were 13 instances of a protest with over 1000 people attending"
di=[i for i in s.split() if i.isdigit()]
print(di)
```

```python
#8.	Given numbers = range(20), produce a list containing the word ‘even’ if a number in the numbers is even, 
#and the word ‘odd’ if the number is odd. Result would look like ‘odd’,’odd’, ‘even’
result = ['even' if n%2 == 0 else 'odd' for n in range(1,21)]
print(result)
```

```python
#9.	Produce a list of tuples consisting of only the matching numbers in these lists list_a = 1, 2, 3,4,5,6,7,8,9, list_b = 2, 7, 1, 12. Result would look like (4,4), (12,12)
list_a = 1, 2, 3,4,5,6,7,8,9 
list_b = 2, 7, 1, 12
l=[(i,j) for i in list_a for j in list_b if i==j]
print(l)
```

```python
#10.	Find all of the words in a string that are less than 4 letters


s="hello people hii"
st=s.split(" ")
s1=list(filter(lambda x:len(x)<4,st))
print((s1))
```
