### 1. Select the Name and Position of each player in the dataframe

```python
df.createOrReplaceTempView("tempview")
i=spark.sql('select name ,position from tempview')
i.show()i=spark.sql('select name ,position from tempview')
i.show()
```

### 1.1 Display the same results from above sorted by the players names

```python
i=spark.sql('select name ,position from tempview')
i.sort('name').show()
```

### 2. Select only the players who belong to a club begining with FC

```python

i=spark.sql('select name from tempview where club like "FC%"')
i.show()
```

### 3. Who is the oldest player in the dataset and how old are they?

Display only the name and age of the oldest player.

```python
j=spark.sql('select name,age from tempview order by age desc')
j.show(1)
```

### 4. Select only the following players from the dataframe:

 - L. Messi
 - Cristiano Ronaldo

```python
i=spark.sql('select name,age from tempview where name like "%L. Messi%" or name like "%Cristiano Ronaldo%"')
i.show()
```

### 5. Can you select the first character from the Release Clause variable which indicates the currency used?

```python
i=df.withColumnRenamed('Release Clause','RC')
i.createOrReplaceTempView("temp")
i.select('RC',i['RC'].substr(1,1)).show(1)
```

### 6. Can you select only the players who are over the age of 40?

```python
i=spark.sql('select name,age from tempview where age>40 ')
i.show()
```
