```python
#line graph
dist_name=['satara','pune','mumbai','nashik']
pop=[60,150,200,70]
plt.plot(dist_name,pop,'r*--',markeredgecolor='green')
plt.xlabel("district")
plt.ylabel("population")
plt.show()
```

```python
#line graph
#values of graph
dist_name=['satara','pune','mumbai','nashik']
pE=[60,150,200,70]
pW=[34,20,250,170]
pN=[60,160,100,200]
pS=[60,450,200,100]
plt.plot(dist_name,pE,'r*--',markeredgecolor='green',label="EAST")
plt.plot(dist_name,pW,color="green",label="WEST")
plt.plot(dist_name,pN,color="yellow",label="NORTH")
plt.plot(dist_name,pS,color='blue',label="SOUTH")
plt.title("population of districts")
plt.xlabel("district")
plt.ylabel("temperature")
#plt.legend()
#plt.legend(loc='upper left')
plt.legend(bbox_to_anchor=(0.75,1.3),ncol=2)
plt.show()
```

```python
#bar plt #width=0.8 (Defalut) alignent plot is at center dimension bottom=0
dist_name=['satara','pune','mumbai','nashik']
t=[32.2,30.5,35.6,28.4]
plt.bar(dist_name,t,width=0.3,color='red')
plt.xlabel("district")
plt.ylabel("temperature")
plt.show()
```

```python
dnames=['satara','pune','mumbai','nashik']
pofE=[60,100,200,90]
pofW=[90,40,50,70]
pofS=[89,90,45,67]
pofN=[70,67,78,89]
bar1=[2,4,6,8]
bar2=[i+0.3 for i in bar1]
bar3=[i+0.3 for i in bar2]
bar4=[i+0.3 for i in bar3]
plt.bar(bar1,pofE,label='East',width=0.3)
plt.bar(bar2,pofW,label='West',width=0.3)
plt.bar(bar3,pofS,label='South',width=0.3)
plt.bar(bar4,pofN,label='North',width=0.3)
plt.xlabel('District names')
plt.ylabel('Population')
plt.legend()
plt.xticks(bar2,dnames)
plt.show()
```

```python
dnames=['Pune','Nashik','Mumbai','satara']
day1=[12,45,23,34]
day2=[12,45,23,34]
day3=[12,59,60,32]
bar1=[2,4,6,8]
bar2=[i+0.3 for i in bar1]
bar3=[i+0.3 for i in bar2]
plt.bar(bar1,day1,label='day1',width=0.3)
plt.bar(bar2,day2,label='day2',width=0.3)
plt.bar(bar3,day3,label='day3',width=0.3)
plt.xlabel('District names')
plt.ylabel('Population')
plt.legend(bbox_to_anchor=(0.7,1.3),ncol=3)
plt.xticks(bar2,dnames)
plt.show()
```

```python
#stacked bar graph
dnames=['Pune','Nashik','Mumbai','satara']
pofE=[60,100,200,90]
pofW=[90,40,50,70]
pofS=[45,89,90,67]
pofN=[60,50,40,80]
bot3=[x+y for x,y in zip(pofE,pofW)]
bot4=[x+y for x,y in zip(bot3,pofS)]
plt.bar(dnames,pofE,color='yellow',width=0.3,label='East')
plt.bar(dnames,pofW,bottom=pofE,color='purple',width=0.3,label='West')
plt.bar(dnames,pofS,bottom=bot3,color='red',width=0.3,label='South')
plt.bar(dnames,pofN,bottom=bot4,color='black',width=0.3,label='North')
plt.xlabel('DIstrict names')
plt.ylabel('Population')
plt.title('population area wise')
plt.legend(bbox_to_anchor=(0.75,1.30),ncol=2)
plt.show()
```

```python

city=['nashik','Pune','Mumbai','satara']
day1=[21,22,30,20]
day2=[12,17,23,24]
day3=[24,23,27,19]
plt.figure(figsize=(8,4))
bot3=[x+y for x,y in zip(day1,day2)]
plt.bar(city,day1,color='yellow',width=0.4,label='Day1')
plt.bar(city,day2,bottom=day1,color='red',width=0.4,label='Day2')
plt.bar(city,day3,bottom=bot3,color='green',width=0.4,label='Day3')
plt.xlabel('City')
plt.ylabel('Temp')
plt.title('temperature')
plt.legend(bbox_to_anchor=(0.75,1.30),ncol=3)
plt.show()


dist_name=['satara','pune','mumbai','nashik']
t=[32.2,30.5,35.6,28.4]
plt.plot(dist_name,t,'r*--',markeredgecolor='green')
plt.xlabel("district")
plt.ylabel("temperature")
plt.show()
```
