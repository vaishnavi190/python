```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dftip=sns.load_dataset('tips')
```

```python
#sea born gives easy ways to create the plot.
#better way of visualization
#in sea born there are different dataset hence it loads the dataset online reporistory...tips,tatainic,breastcancer and many more
```

```python
dftip#is a dataframe which has all this columns
```

```python
#to check the realationship (scatter plot),positive corelation
sns.relplot(x='total_bill',y='tip',data=dftip)
plt.show()
#total_bill is independent(predictor)variable where as tip is dependent
```

```python
#sex,smoke,day,time.size-categorical
#totalbill,tip-continuse
sns.relplot(x='total_bill',y='tip',hue='sex',data=dftip)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',data=dftip)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',style='day',data=dftip)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',style='day',size='size',data=dftip)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',style='smoker',col='day',col_wrap=2,data=dftip)
plt.show()
```

```python
dftip.corr()
```

```python
sns.heatmap(dftip.corr())
plt.show()
```

```python
sns.pairplot(dftip)
plt.show()
```
